Note that majority value if "Office" in WorkLoc

replace all empty rows in column WorkLoc with the value of "Office"

```python
workloc_highest = 'Office'

missing_data = df.isnull()
#print(missing_data.head(5))

print('\nValue counts for missing data in WorkLoc:\n')
print( missing_data['WorkLoc'].value_counts())


print('\nHere are the first 10 rows missing values for WorkLoc:')
print(df[missing_data['WorkLoc']][['Respondent','WorkLoc']].head(10))

df['WorkLoc'].replace(np.nan, workloc_highest, inplace=True)
```

```python
print('\nNew value counts for missing data in WorkLoc:\n')
print(missing_data['WorkLoc'].value_counts())
```

##### No 'True' values means that we have eliminated all empty WorkLoc values.

## Normalizing data

There are 2 cols "CompFreq" and "CompTotal" shows how often a developer is paid (Yearly, Monthly, Weekly). and how much the developer is paid per Year, Month, or Week depending upon his/her "CompFreq".

will create a new column called 'NormalizedAnnualCompensation' which contains the 'Annual Compensation' irrespective of the 'CompFreq'.

```python
print(df['CompFreq'].value_counts())
```

```python
df.loc[df['CompFreq'] == 'Yearly', 'NormalizedAnnualCompensation']  = 1  * df['CompTotal']
df.loc[df['CompFreq'] == 'Monthly', 'NormalizedAnnualCompensation'] = 12 * df['CompTotal']
df.loc[df['CompFreq'] == 'Weekly', 'NormalizedAnnualCompensation']  = 52 * df['CompTotal']


df[['CompTotal','CompFreq','NormalizedAnnualCompensation']].head(20)

df['NormalizedAnnualCompensation'].median()
```

## using MinMax scaler

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_comp_total = scaler.fit_transform(df["CompTotal"].values.reshape(-1, 1))
df["ScaledCompTotal"] = scaled_comp_total
```

```python
print(df["ScaledCompTotal"])
```

```python
scaled_workhrs_total = scaler.fit_transform(df["WorkWeekHrs"].values.reshape(-1, 1))
df["ScaledWorkWkHrs"] = scaled_workhrs_total
```
